## Scrapping Wiki Louvre

```python
import requests
import json

url_api = "https://fr.wikipedia.org/w/api.php"

parametres = {
    "action": "query",
    "prop": "extracts",
    "exintro": True,
    "explaintext": True,
    "titles": "Musée du Louvre",
    "format": "json",
}

headers = {"User-Agent": "Scrapping_wiki"}

try:
    reponse = requests.get(url_api, params=parametres, headers=headers)
    reponse.raise_for_status()
    donnees = reponse.json()

    # Sauvegarde du dictionnaire complet dans un fichier .json
    with open("scrapping/wiki/scrapping_louvre.json", "w", encoding="utf-8") as f:
        # indent=4 : crée des sauts de ligne et décalages pour que ce soit lisible
        # ensure_ascii=False : indispensable pour ne pas transformer les "é" en "\u00e9"
        json.dump(donnees, f, indent=4, ensure_ascii=False)

    print("Le fichier 'scrapping_louvre.json' a été créé avec succès.")

except Exception as e:
    print(f"Erreur : {e}")
```

```python
import ollama
import json

# 1. Charger les données scrapées
with open("scrapping/wiki/scrapping_louvre.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# On récupère le texte (en supposant la structure de ton code précédent)
pages = data['query']['pages']
page_id = list(pages.keys())[0]
texte_wiki = pages[page_id]['extract']

# 2. Envoyer une question à Ollama sur ce texte
# On limite le texte envoyé pour ne pas saturer la mémoire si l'article est géant
prompt = f"Voici un extrait de Wikipédia sur le Louvre : {texte_wiki[:2000]}. Peux-tu me lister les 3 dates clés citées ?"

response = ollama.chat(model='llama3', messages=[
  {
    'role': 'user',
    'content': prompt,
  },
])

print("--- Analyse de l'IA ---")
print(response['message']['content'])
```
