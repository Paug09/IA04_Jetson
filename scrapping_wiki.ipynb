```python
import requests

# L'URL de l'API de Wikipédia en français
url_api = "https://fr.wikipedia.org/w/api.php"

# Les paramètres de notre recherche
parametres = {
    "action": "query",
    "prop": "extracts",    # On veut extraire du texte
    "exintro": True,       # Uniquement l'introduction
    "explaintext": True,   # En texte brut (pas de balises HTML)
    "titles": "Python (langage)", # Le titre exact de la page
    "format": "json"       # Le format de réponse souhaité
}

# Wikipédia demande d'avoir un "User-Agent" personnalisé pour savoir qui utilise son API
headers = {"User-Agent": "MonProjetDeScraping/1.0 (votre_email@exemple.com)"}

reponse = requests.get(url_api, params=parametres, headers=headers)
donnees = reponse.json()

# Naviguer dans le dictionnaire JSON pour trouver le texte
pages = donnees['query']['pages']
for page_id in pages:
    print(pages[page_id]['extract'])
```
