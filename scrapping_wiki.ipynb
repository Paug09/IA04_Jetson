## Scrapping Wiki Louvre

```python
import requests
import json

url_api = "https://fr.wikipedia.org/w/api.php"

parametres = {
    "action": "query",
    "prop": "extracts",
    "exintro": True,
    "explaintext": True,
    "titles": "Musée du Louvre",
    "format": "json",
}

headers = {"User-Agent": "Scrapping_wiki"}

try:
    reponse = requests.get(url_api, params=parametres, headers=headers)
    reponse.raise_for_status()
    donnees = reponse.json()

    # Sauvegarde du dictionnaire complet dans un fichier .json
    with open("scrapping/wiki/scrapping_louvre.json", "w", encoding="utf-8") as f:
        # indent=4 : crée des sauts de ligne et décalages pour que ce soit lisible
        # ensure_ascii=False : indispensable pour ne pas transformer les "é" en "\u00e9"
        json.dump(donnees, f, indent=4, ensure_ascii=False)

    print("Le fichier 'scrapping_louvre.json' a été créé avec succès.")

except Exception as e:
    print(f"Erreur : {e}")
```
